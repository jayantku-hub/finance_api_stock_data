```python
import requests
import pandas as pd
import time
```

```python
symbols = ["APPL", "MSFT", "GOOGL", "TSLA", "AMZN"]
all_stocks = []
```

```python
# Import required libraries

API_KEY = "UKEML6XYDOORPBK3"

# This is the address of the Alpha Vantage server where we send requests to get financial data.
url = "https://www.alphavantage.co/query"

# Loop through multiple stock symbols
for symbol in symbols:

# Creates a dictionary of parameters to send to the API

    params = {
        "function": "GLOBAL_QUOTE",  #tells the API we want current stock price
        "symbol": symbol,            #tells which stock (AAPL, MSFT, etc.)
        "apikey": API_KEY            #sends your identity to the API
    }


    response = requests.get(url, params=params)  #Sends a GET request to the API with the given parameters.
    data = response.json()                       #Converts the API response into a Python dictionary.

    quote = data.get("Global Quote", {})         #Tries to extract "Global Quote" from the response.


    # Extract required stock fields
    if quote:
        stock_info = {
            "Symbol": quote.get("01. symbol"),
            "Price": quote.get("05. price"),
            "Previous_Close": quote.get("08. previous close"),
            "Change_Percent": quote.get("10. change percent")
        }
        all_stocks.append(stock_info)

    time.sleep(12)  #  API rate limit



```

```python
df = pd.DataFrame(all_stocks)
df
```
